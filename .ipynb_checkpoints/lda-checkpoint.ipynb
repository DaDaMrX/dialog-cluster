# LDA主题聚类

## 1. 数据预处理
- Combine utterance
- Delete words

```python
import re
import jieba

raw_file = 'chat-short-20w.txt'
clean_file = 'chat-short-clean-20w.txt'
stop_wrods_file = 'stop_words.txt'

def substitute(sent):
    exps = [
        r'#E-\w\[数字x\]|~O\(∩_∩\)O/~',
        r'http[s]?://[a-zA-Z0-9|\.|/]+',
        r'http[s]?://[a-zA-Z0-9\./-]*\[链接x\]',
        r'\[ORDERID_[0-9]+\]',
        r'\[日期x\]',
        r'\[时间x\]',
        r'\[金额x\]',
        r'\[站点x\]',
        r'\[数字x\]',
        r'\[地址x\]',
        r'\[姓名x\]',
        r'\[邮箱x\]',
        r'\[电话x\]',
        r'\[商品快照\]',
        r'<s>',
        r'\s+',
        r'[a-z|0-9]+'
        "[\s+\.\!\/_,$%^:*(+\"\')]+",
        "[+——()?:【】‘’“”`！，。？、~@#￥%……&*（）]+"
    ]
    for exp in exps:
        sent = re.sub(exp, ' ', sent)
    return sent

# Read raw file
with open(raw_file, encoding='utf-8') as f:
    corpus = f.read()

# Combine 
corpus = corpus.strip().split('\n\n')
for i in range(len(corpus)):
    corpus[i] = corpus[i].strip().split('\n')
    for j in range(len(corpus[i])):
        corpus[i][j] = corpus[i][j].strip()[2:]
    corpus[i] = ' '.join(corpus[i])

# Delete
corpus = list(map(substitute, corpus))

# Tokenize
corpus = list(map(jieba.cut, corpus))

# Stop words
with open(stop_wrods_file, encoding='utf-8') as f:
    stop_words = f.read().strip().split('\n')

# Filter
for i in range(len(corpus)):
    tokens = []
    for token in corpus[i]:
        token = token.strip()
        if len(token) > 1 and token not in stop_words:
            tokens.append(token)
    corpus[i] = tokens

# Combine
corpus = list(map(lambda x: ' '.join(x), corpus))

# Show
print(len(corpus), 'conversations')
print(corpus[0])
```

## 2. 获取词频矩阵

```python
from sklearn.feature_extraction.text import CountVectorizer

min_df = 10
max_df = 0.1
max_features = 1000

vectorizer = CountVectorizer(min_df=min_df,
                             max_df=max_df,
                             max_features=max_features)
tf = vectorizer.fit_transform(corpus)

vocabulary = vectorizer.get_feature_names()
print(len(vocabulary), 'words.')
```

## 3. LDA主题模型

```python
from sklearn.decomposition import LatentDirichletAllocation

n_topics = 20
max_iter=50

learning_method='online'
learning_decay=0.7
learning_offset=10.0
batch_size=128

evaluate_every=5
perp_tol=0.001

n_jobs=4
random_state=0
verbose=1

lda = LatentDirichletAllocation(n_components=n_topics,
                                max_iter=max_iter,
                                learning_method=learning_method,
                                learning_decay=learning_decay,
                                learning_offset=learning_offset,
                                batch_size=batch_size,
                                evaluate_every=evaluate_every,
                                perp_tol=perp_tol,
                                random_state=random_state,
                                verbose=verbose,
                                n_jobs=n_jobs)
doc_topic = lda.fit_transform(tf)
```

## 4. 结果

### 4.1 每个主题的前n个关键词

```python
n_top_words = 20
vocabulary = vectorizer.get_feature_names()
for topic_idx, topic in enumerate(lda.components_):
    print('Topic #%d:' % topic_idx)
    words = ' '.join([vocabulary[i] for i in topic.argsort()[:-1-n_top_words:-1]])
    print(words, end='\n\n')
```
